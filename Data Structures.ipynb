# Dictionary

```python
products = {"product name": "colgate", "price": 100, "quantity": 10}
print("product Details:", products)
```

```python
products = {"product name": "colgate", "price": 100, "quantity": 10}
products ["quantity"] = 5       
print("After updating quantity:", products)
```

```python
products = {"product name": "colgate", "price": 100, "quantity": 10}
print(products.keys())
```

```python
products = {"product name": "colgate", "price": 100, "quantity": 10}
for key,value in products.items():
    print(f"key:{key}, Value:{value}")
```
