# User define function

```python
def evenorodd(num):
    
  if num %2==0:
        return "even"
  else:
        return "odd"
```

```python
evenorodd(10)
```

```python
evenorodd(165)
```

```python
def factorial(n):
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers.")
    if n == 0:
        return 1
    else:
        n>=1
        return n*factorial(n-1)
```

```python
print(factorial(8))
```

```python
print(factorial(4))
```

```python
def maximum(num1,num2,num3):
    if num1>=num2 and num1>=num3:
        return num1
    elif num2>=num1 and num2 >= num3:
        return num2
    else:
        return num3
```

```python
maximum(44,23,86)
```

 # Lambda Function

```python
multiply =lambda a,b:a*b
print(multiply(4,5))
```

```python
print(multiply(8,32))
```

```python
a=[1,2,3,4,5,6,7,8,10]
even=list(filter(lambda a:a%2==0,a))
print(even)
```

```python
str=["hello","world"]
uppercase=list(map(lambda string:string.upper(),str))
print(uppercase)
```
